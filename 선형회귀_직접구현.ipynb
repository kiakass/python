```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3])
y = np.array([2, 4, 6])
```

```python
# 최대 반복 횟수
num_epoch = 2000

# 학습율 (learning_rate)
learning_rate = 0.01
errors = []
# random 한 값으로 w, b를 초기화 합니다.
# w = np.random.uniform(low=-1.0, high=1.0)
# b = np.random.uniform(low=-1.0, high=1.0)
[w,b]=[0,0]


for epoch in range(num_epoch):
    
    #hypothesis
    y_hat = x * w + b
    
    # w, b update
    w = w - learning_rate * 2*((y_hat - y) * x).mean()
    b = b - learning_rate * 2*(y_hat - y).mean()

    #loss : Mean square error
    error = ((y_hat - y)**2).mean()

    if error < 0.00006:
        break

    errors.append(error)
    
    # 그래프 출력
    if epoch % 500 == 0: # step 마다 그래프 출력
        print("step: %3d, w = %.5f, b = %.5f, cost = %.5f" % (epoch,w,b,error))
        plt.plot(x,y,'o', label='step = {}'.format(epoch))
        plt.plot(x,w*x+b)
        plt.legend()
        plt.show()

print("----" * 15)
print("final result : step: %3d, w = %.2f, b = %.2f, cost = %.5f" %(epoch, w, b, error))
```
